```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r'D:\Chrome Downloads\smoking_health_data_final.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['cigs_per_day'], inplace=True)
```

```python
df.dropna(subset=['chol'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
# Distribution of Age

plt.hist(df['age'], bins=20)
plt.title('Distribution by Age')
plt.xlabel('Frequency')
plt.ylabel('Age')
plt.show()
```

```python
df['age'].mode()
```

```python
# Age vs. Heart Rate
plt.scatter(df['age'], df['heart_rate'], alpha=0.5)
plt.title('Age vs. Heart Rate')
plt.xlabel('Age')
plt.ylabel('Heart Rate')
plt.show()
```

```python
# Boxplot of Cigarettes Per Day
sns.boxplot(df['cigs_per_day'])
plt.title('Distribution of Cigarettes Per Day')
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Cigarettes')
plt.show()
```

```python
# Scatterplot of Cholestrol vs. Age
plt.scatter(df['age'], df['chol'])
plt.title('Age vs. Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
sns.boxplot(x='sex', y='chol', data=df)
plt.title('Cholestrol Levels by Sex')
plt.xlabel('Sex')
plt.ylabel('Cholesterol')
plt.show()
```

```python
# Average Heart Rate by Smoking Status
avg_heart_rate = df.groupby('current_smoker')['heart_rate'].mean()
sns.barplot(data=avg_heart_rate, palette=['blue', 'red'])
plt.title('Average Heart Rate by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Average heart Rate')
plt.xticks([0,1], ['Non-Smoker', 'Smoker'], rotation=0)
plt.show()
```

```python
# Average Cholesterol by Smoking Status
avg_chol = df.groupby('current_smoker')['chol'].mean()
sns.barplot(data=avg_chol, palette=['blue', 'red'])
plt.title('Average Cholesterol by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Average Cholesterol')
plt.xticks([0,1], ['Non-Smoker', 'Smoker'], rotation=0)
plt.show()
```

```python
# Heart Rate by Gender
sns.boxplot(x=df['sex'], y=df['heart_rate'])
plt.xlabel('Gender')
plt.ylabel('Heart Rate')
plt.title('Heart Rate Distribution by Gender')
plt.show()
```

```python
# Cigarettes Per Day vs. Cholesterol
plt.scatter(df['cigs_per_day'], df['chol'])
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Cholesterol')
plt.title('Cigarettes Per Day vs. Cholesterol')
plt.show()
```

```python
# Age Group Distribution
age_groups = pd.cut(df['age'], bins=range(20, 81, 10))
age_groups.value_counts().sort_index().plot(kind='bar')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Age Groups')
plt.xticks(rotation=45)
plt.show()
```

```python
# Cholesterol Levels Distribution
plt.hist(df['chol'])
plt.xlabel('Cholesterol Levels')
plt.ylabel('Frequency')
plt.title('Distribution of Cholesterol Levels')
plt.show()
```
